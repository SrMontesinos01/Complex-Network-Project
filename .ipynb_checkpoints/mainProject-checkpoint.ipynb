# Reading the Network

```python
def read_file(root):
    path_to_zipped_file = root

    # open zip with network files
    with zipfile.ZipFile(path_to_zipped_file) as z:
        # open file for the edges 
        with z.open('edges.csv') as f:
            df = pd.read_csv(f)
            df = df.rename(columns={"# source": " source"})
            for col in df.columns:
                df = df.rename(columns={col: col[1:]})
            # create network from edge list (call the proper network constructor, e.g., DiGrpah if directed)
            G = nx.Graph(df)
            
        # open file for node properties and add them to the graph
        with z.open('nodes.csv') as f:
            df = pd.read_csv(f)
            df = df.rename(columns={"# index": " id"})
            for col in df.columns:
                df = df.rename(columns={col: col[1:]})
            for att in list(df.columns[1:]):
                att_dict = pd.Series(df[att].values,index=df.id).to_dict()
                nx.set_node_attributes(G,att_dict,att)
    return G
```

```python
G = read_file("network_9.zip")

for edge in G.edges():
    if '_graphml_edge_id' in G.edges[edge]:
        del G.edges[edge]['_graphml_edge_id']

for edge in G.edges():
    if 'weight' in G.edges[edge]:
        del G.edges[edge]['weight']

for node in G.nodes():
    if '_graphml_vertex_id' in G.nodes[node]:
        del G.nodes[node]['_graphml_vertex_id']

components = list(nx.connected_components(G))
largest_component = max(components, key=len)
G = G.subgraph(largest_component).copy()
```

# Part 1 

## a) Number of nodes, links, network density

The total amounts of nodes is quite different from the one that is shown in the corresponding webpage. This is due to the fact that we are just using the main component of the network. The same happens with the number of edges.

```python
def degree_ave(gr):
    """
    Given a graph "gr", calculates the average of the degree
    """
    kArr = np.array([k for node, k in G.degree()])
    return np.sum(kArr)/ len(kArr)

def degree_var(gr):
    """
    Given a graph "gr", calculates the variance of the degree
    """
    kArr = np.array([k for node, k in G.degree()])
    mean = degree_ave(gr)
    return np.sum((kArr - mean)**2) / len(kArr)
```

```python
n = len(G.nodes()) # Number of nodes
m = len(G.edges()) # Number of edges
d = nx.density(G)  # d =  2m/n(n-1) (Graph Density)

print("Number of nodes:", n)
print("Number of edges/links:", m)
print("Network Density:", d)
print("Mean degree; std:", degree_ave(G), ";", np.sqrt(degree_var(G)))
```

## b) Average clustering coefficient of the nodes (Watts-Strogatz clustering coefficient)

This coefficient is computed as:
$$
\bar C = \frac{1}{n}\sum _i C_i = \frac{1}{n}\sum _i \frac{2t_i}{k_i(k_i -1)}
$$
where $t_i$ is the number of transitive relations of the node $i$. See at networkx documentation: https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.cluster.average_clustering.html


```python
aveClusterG = nx.average_clustering(G)
print("Average clustering coefficient of the nodes (Watts-Strogatz clustering coefficient)", aveClusterG)
```

## c) Global clustering coefficient (Newman transitivity)

This coefficient is computed as:
$$
C = \frac{3|C_3|}{|P_2|}, \;\;\;\; |C_3| = \frac{1}{6}\text{tr}(A^3)\;\;\;\; |P_2| = \frac{1}{2}\sum_{i = 1}^N k_i(k_i - 1)
$$
where $|C_3|$ is the total number of triangles of the network, $|P_2|$ the total number of paths of lengh two of the network.
https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.cluster.transitivity.html#networkx.algorithms.cluster.transitivity

```python
globalCluster = nx.transitivity(G)
print("Average clustering coefficient of the nodes (Watts-Strogatz clustering coefficient)", globalCluster)
```

### Fitting

```python
def fit_and_Plots(distPdf, distCdf, initGuess, **kargs):
    """
    This functions takes the PDF and CDF of a certain distribution function and performs a fit 
    using the "initGuess". The fit is done for the CDF
    """
    x= np.linspace(min(kUniqList) - 20, max(kUniqList), 100)
    cdfInit = distCdf(x, **kargs)
    pdfInit = distPdf(x, **kargs)
    
    popt, pcov = sp.optimize.curve_fit(distCdf, kUniqList, cumFreqList, p0=initGuess)
    perr = np.sqrt(np.diag(pcov)) # compute errors
    cdfFit = distCdf(kUniqList, *popt)
    pdfFit = distPdf(kUniqList, *popt)

    ms = 20
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))
    ax1 = axs[0]
    ax1.plot(x, pdfInit, label='Init Guess', alpha = 0.4, color='g')
    ax1.plot(kUniqList, pdfFit, label='Fit', alpha = 0.6, color='r')
    ax1.scatter(kUniqList, freqList, alpha=0.75, color='blue', label="Data")
    
    
    ax2 = axs[1]
    ax2.scatter(x, cdfInit, label='Init Guess', alpha = 0.4, color='g',s = ms, marker = 's')
    ax2.plot(kUniqList, cdfFit, alpha = 0.6, color='r')
    ax2.scatter(kUniqList, cumFreqList, alpha=0.75, color='blue')
    plt.legend()

    # cdfFit = distCdf(x, *popt)
    # pdfFit = distPdf(x, *popt)

    # ax1.plot(x, pdfFit, 'yo-', label='Fit', alpha = 0.5)
    # ax2.plot(x, cdfFit, 'yo-', label='Fit', alpha = 0.5)
    # ax1.set_xscale("log")
    # ax1.set_yscale("log")
    # ax1.set_ylim(0.001, 0.2)

    #ax2.set_xscale("log")
    #ax2.set_yscale("log")
    # ax1.set_xlim(0.1, 220)
    return pdfFit, cdfFit, popt, perr
```

```python
def fit_and_Plots_v2(distPdf, distCdf, initGuess, **kargs):
    """
    This functions takes the PDF and CDF of a certain distribution function and performs a fit 
    using the "initGuess". The fit is done for the PDF
    """
    x= np.linspace(min(kUniqList) -20, max(kUniqList), 100)
    cdfInit = distCdf(x, **kargs)
    pdfInit = distPdf(x, **kargs)
    
    popt, pcov = sp.optimize.curve_fit(distPdf, kUniqList, freqList, p0=initGuess)
    # popt, pcov = sp.optimize.curve_fit(distPdf, kUniqList, freqList)
    perr = np.sqrt(np.diag(pcov)) # compute errors

    #shape, scale = popt
    cdfFit = distCdf(kUniqList, *popt)
    pdfFit = distPdf(kUniqList, *popt)

    fig, axs = plt.subplots(1, 2, figsize=(8, 4))
    ax1 = axs[0]
    ax1.plot(x, pdfInit, 'go-', label='Init Guess', alpha = 0.5)
    ax1.plot(kUniqList, pdfFit, 'ro-', label='Fit', alpha = 0.5)
    ax1.plot(kUniqList, freqList, alpha=0.75, color='blue', label="Data")
    
    ax2 = axs[1]
    ax2.plot(x, cdfInit, 'go-', label='Init Guess', alpha = 0.5)
    ax2.plot(kUniqList, cdfFit, 'ro-', label='Fit', alpha = 0.5)
    ax2.plot(kUniqList, cumFreqList, alpha=0.75, color='blue')
    plt.legend()

    # cdfFit = distCdf(x, *popt)
    # pdfFit = distPdf(x, *popt)

    # ax1.plot(x, pdfFit, 'y-', label='Fit', alpha = 0.5)
    # ax2.plot(x, cdfFit, 'yo-', label='Fit', alpha = 0.5)
    return pdfFit, cdfFit, popt, perr
```

#### Gamma (x>= 0)

## f) List the top 25 most central nodes
According to each of the following centrality measures:
degree, closeness, betweenness, eigenvector, Katz index, PageRank and subgraph
centrality. Compare the rankings and draw conclusions.

The slides corresponding to "Reading 5"

```python
def topCent(graph, cenMeasure, numNodes=25, **kwargs):
    centDict = cenMeasure(graph, **kwargs) # Obtains de nx.centrality(G) for whatever centrality Measure
    return sorted(centDict, key=centDict.get, reverse=True)[:numNodes]
    
# Degree
topDegCent = topCent(G, nx.degree_centrality)
print("topDegCent: ", topDegCent)

# Closeness
topCloCent = topCent(G, nx.closeness_centrality)
print("topCloCent: ", topCloCent)

# Betweenness
topBetCent = topCent(G, nx.betweenness_centrality)
print("topBetCent: ", topBetCent)

# Eigevector
topEigCent = topCent(G, nx.eigenvector_centrality)
print("topEigCent: ", topEigCent)

# Katz
A = nx.to_numpy_array(G)
eigvals, eigvect = np.linalg.eigh(A) # Obtains eigenvalues and eigenvectors of A
a = 1/max(eigvals)
print("a:", a)
topKatCent = topCent(G, nx.katz_centrality, numNodes = 25, alpha = a/2) 
print("topKatCent: ", topKatCent)

# PageRank
topPagCent = topCent(G, nx.pagerank)
print("topPagCent: ", topPagCent)

# Subgraph
topSubCent = topCent(G, nx.subgraph_centrality)
print("topSubCent: ", topSubCent)
```

## g) Report the degree assortativity of this network.

The degree assortativity is computed as ("Reading 4"): 
$$
r=\frac{\frac{1}{m} \sum_{(i, j) \in E} k_i k_j-\left(\frac{1}{2 m} \sum_{(i, j) \in E}\left(k_i+k_j\right)\right)^2}{\frac{1}{2 m} \sum_{(i, j) \in E}\left(k_i^2+k_j^2\right)-\left(\frac{1}{2 m} \sum_{(i, j) \in E}\left(k_i+k_j\right)\right)^2}
$$

The degree assortativity of a graph measures the tendency of nodes with similar degrees to be connected to each other. In other words, it quantifies the correlation between the degrees of connected nodes in a network. A positive assortativity coefficient indicates that nodes with similar degrees tend to be connected, while a negative coefficient suggests that nodes with different degrees are more likely to be connected.

Degree assortativity is a way to analyze the structure of a network and understand how nodes with different degrees interact. Networks with high degree assortativity often have a modular structure, where nodes with similar degrees form clusters or communities within the larger network.

The degree assortativity coefficient is calculated using various methods, with one common measure being the Pearson correlation coefficient applied to the degrees of connected nodes. The formula for the assortativity coefficient can vary depending on the context and the specific definition used by researchers.

```python
assort = nx.degree_assortativity_coefficient(G)
print("Degree assortativity of this network:", assort) # Seems to be the same as in class
```

```python
def assortDegree(graph):
    m = len(graph.edges) # Total number of edges in the network
    sk, sk2, sk3 = 0, 0, 0
    
    # Iteration over every edge of the graph
    for edge in graph.edges:
        # Obtain the Degree of the nodes that are connected by de edge
        k1, k2 = graph.degree(edge[0]), graph.degree(edge[1]) 
        
        sk += k1 + k2
        sk2 += k1*k1 + k2*k2
        sk3 += k1*k2
        
    num = sk3/m - (sk/(2*m))**2
    den = sk2/(2*m) - (sk/(2*m))**2
    return num/den
```

```python
def bipartIndex(graph):
    A = nx.to_numpy_array(graph) # Returns the graph adjacency matrix as a NumPy array. 
    
    eigvals, eigvect = np.linalg.eigh(A) # Obtains eigenvalues and eigenvectors of A

    val = np.trace(sp.linalg.expm(-A)) / np.trace(sp.linalg.expm(A))
    # val = num/den
    return val

bipartIndex(G)
```

## Summarizing

```python
def basicAnalysis(graph):
    """
    Given a "graph", computes:
        b) Average clustering coefficient of the nodes (Watts-Strogatz clustering coefficient)
        c) Global clustering coefficient (Newman transitivity)
        d) Average path length and diameter of the network
        e) Degree distribution of the network
        g) Degree assortativity of this network
        h) Bipartite Index

    Returns:
        - np.array with b)-d) and g)-h) [total of 6 measures]
        - list "distInfor" with the k unique values, the pdf and cdf
    """

    aveCluster = nx.average_clustering(graph) # b)
    globalCluster = nx.transitivity(graph) # c)
    avePathLength = nx.average_shortest_path_length(graph) # d)
    diamG = nx.diameter(graph) # d)

    # Compute degree distribution
    freqList = np.array(nx.degree_histogram(graph)) # "A list of frequencies of degrees. The degree values are the index in the list." (Documentation)
    freqList = freqList / n # Normalize
    cumFreqList = np.cumsum(freqList) # Cumulative
    kUniqList = range(len(freqList))
    distInfo = [kUniqList, freqList, cumFreqList]

    assort = nx.degree_assortativity_coefficient(graph) # g)
    bipart = bipartIndex(graph) # h)

    return np.array([aveCluster, globalCluster, avePathLength, diamG, assort, bipart]), distInfo
```

```python
def plotDistPart2(kUniqList, freqList, cumFreqList, distList):
    pdfs = []
    cdfs = []
    kVal = [] 
    for distInfo in distList:
        # print(distInfo[0])
        kVal.append(distInfo[0]) # Index 0 correspond to the K values
        pdfs.append(distInfo[1]) # Index 1 correspond to the pdf
        cdfs.append(distInfo[2]) # Index 2 correspond to the cdf
        
    # Search for the maximum krange    
    kValAux = kVal.copy()
    kValAux.append(kUniqList)
    kMax = max(kValAux, key = len)
    # print(kMax)
    # print(len(kMax) - len(kVal[0]))
    
    # Add zeros to the pdf and ones to the cdf so  all of them have the same lenght as the distribution of our network G (kMax)
    # len(kMax) - len(kRange) is the number of 0's or 1's that needs to be added
    pdfs = [np.pad(distArr, (0, len(kMax) - len(kRange)), 'constant') for distArr, kRange in zip(pdfs, kVal)]
    cdfs = [np.pad(distArr, (0, len(kMax) - len(kRange)), 'constant', constant_values=1) for distArr, kRange in zip(cdfs, kVal)]
    freqListZeros =  np.pad(freqList, (0, len(kMax) - len(kUniqList)), 'constant')
    cumFreqListOnes = np.pad(cumFreqList, (0, len(kMax) - len(kUniqList)), 'constant', constant_values=1)
    
    #Computing the average of each distribution
    # "i" runs over all the points in the distribution
    # "pdf" runs over all the np.arrays that are in pdfs
    avePdf = np.array([np.mean([pdf[i] for pdf in pdfs]) for i in range(len(pdfs[0]))])
    stdPdf = np.array([np.std([pdf[i] for pdf in pdfs]) for i in range(len(pdfs[0]))])
    
    aveCdf = np.array([np.mean([cdf[i] for cdf in cdfs]) for i in range(len(cdfs[0]))])
    stdCdf = np.array([np.std([cdf[i] for cdf in cdfs]) for i in range(len(cdfs[0]))])
    
    # pdfSup = avePdf + stdPdf
    # pdfInf = avePdf - stdPdf
    # cdfSup = aveCdf + 5*stdCdf
    # cdfInf = aveCdf - 5*stdCdf

    return kMax, avePdf, stdPdf, aveCdf, stdCdf, freqListZeros, cumFreqListOnes
```

## Small world property

```python
def errorDiv(x, y, ex, ey):
    result = x / y
    
    dzdx = 1 / y
    dzdy = -x / (y ** 2)
    
    error_z = np.sqrt((dzdx * ex)**2 + (dzdy * ey)**2)
    
    return result, error_z
```

## Mutual information

```python
from sklearn.metrics.cluster import normalized_mutual_info_score as NMI
```

```python
def labels_from_communities(communities):
    '''
    This function gives a vector with the labels of nodes in a partition given by communities
    '''
    N = 0
    for comm in communities:
        N += len(comm)
    labels = N*[[]]
    i = 0
    for comm in communities:
        for node in comm:
            labels[node] = i
        i += 1
    return labels
```

```python
def edgeType(gr, name):
    edgeTypeDict = dict()
    for edge in gr.edges():
        type_1 = gr.nodes[edge[0]][name]
        type_2 = gr.nodes[edge[1]][name]
        
        if type_1 == type_2:
            edgeTypeDict[edge] = 'Same'
        else:
            edgeTypeDict[edge] = 'Mixed'
            
    return edgeTypeDict
        
def prepare_cyto(G, name):
    
    # Edge attribute to see if the edge connects nodes from the same community
    edgeLou = edgeType(G, "comLou")
    nx.set_edge_attributes(G,edgeLou,'edgeLou')

    edgeMod = edgeType(G, "comMod")
    nx.set_edge_attributes(G,edgeMod,'edgeMod')

    edgeLab = edgeType(G, "comLab")
    nx.set_edge_attributes(G,edgeLab,'edgeLab')

    edgeInf = edgeType(G, "comInf")
    nx.set_edge_attributes(G,edgeLab,'edgeInf')
    
    root = './results/' + name +'.gml'
    root = name +'.gml'
    nx.write_gml(G, root)
```
